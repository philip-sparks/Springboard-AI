```python
# The BigQuery client library provides a cell magic, 
# %%bigquery, which runs a SQL query and returns the results as a Pandas DataFrame.
%%bigquery france_subset
SELECT SQLDATE, Actor1Name, SOURCEURL, MonthYear, EventRootCode, GoldsteinScale, NumMentions, AvgTone, ActionGeo_CountryCode, ActionGeo_Lat, ActionGeo_Long
FROM `gdelt-bq.full.events`
WHERE ActionGeo_CountryCode = 'FR' AND MonthYear > 201601
LIMIT 16000
```
